```python
from stable_baselines3.ppo import CnnPolicy
```

```python
from stable_baselines3 import PPO
```

```python
from pettingzoo.butterfly import pistonball_v6
```

```python
import supersuit as ss
```

```python
env = pistonball_v6.parallel_env(n_pistons=20, time_penalty=-0.1, 
                                 continuous=True, random_drop=True, random_rotate=True, ball_mass=0.75, 
                                 ball_friction=0.3, ball_elasticity=1.5, max_cycles=125)
```

```python
help(env.observation_space)
```

```python
env = ss.color_reduction_v0(env, mode='B')
```

```python
env = ss.resize_v1(env, x_size=84, y_size=84)
```

```python
env = ss.frame_stack_v1(env, 3)
```

```python
env = ss.pettingzoo_env_to_vec_env_v1(env)
```

```python
env = ss.concat_vec_envs_v1(env, 8, num_cpus=4, 
                            base_class='stable_baselines3')
```

```python
model = PPO(CnnPolicy, env, verbose=3, gamma=0.95, n_steps=256, 
            ent_coef=0.0905168, learning_rate=0.00062211, vf_coef=0.042202, 
            max_grad_norm=0.9, gae_lambda=0.99, n_epochs=5, clip_range=0.3, batch_size=256)
```

```python
model.learn(total_timesteps=2_000_000)
```

```python
model.save("policy")
```

```python
env = pistonball_v6.env()
env = ss.color_reduction_v0(env, mode='B')
env = ss.resize_v1(env, x_size=84, y_size=84)
env = ss.frame_stack_v1(env, 3)
```

```python
model = PPO.load("policy")
```

```python
env.reset()
for agent in env.agent_iter():
    obs, reward, done, info = env.last()
    act = model.predict(obs, deterministic=True)[0] if not done else None
    env.step(act)
    env.render()
```

```python
env.close()
```
