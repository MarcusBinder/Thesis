```python
A = 1

folder = 'windfarms'
files = os.listdir(folder)
with np.load(os.path.join(folder, files[A])) as data:
    Farm_Layout = data[files[A].split(".")[0]]
X_i = Farm_Layout[0:-1:2]
Y_i = Farm_Layout[1::2]

plt.scatter(X_i, Y_i)
plt.title(files[A].split("_")[0] + " layout")
plt.show()
```
